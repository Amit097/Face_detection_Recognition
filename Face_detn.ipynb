# Let's import necessary Library

```python
from mtcnn.mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('./face_detn1.jpg')
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.show()
```

```python
detector=MTCNN()
boundary_locn=detector.detect_faces(img1)
```

Detect all the faces from given image and draw the rectangle boundary across the face.

```python
print(boundary_locn[0])
for i in range(len(boundary_locn)):
    x1,y1,x2,y2=boundary_locn[i]['box']
    print(x1,x2,y1,y2)
    cv2.rectangle(img1, (x1, y1), (x1+x2, y1+y2), (255,0,0), 2)

plt.figure(figsize=(15,10))
plt.imshow(img1)
plt.show()
```

From the given image crop the all faces and save into Img folder.

```python
import numpy as np
a=[]
for i in range(len(boundary_locn)):
    loc=boundary_locn[i]['box']
    #loc1=[int(x*1) for x in loc]
    x1,y1,x2,y2=loc
    plt.imshow(img[y1:y1+y2,x1:x1+x2])
    filename = "./Img/face_detn1_%d.jpg"%i
    cv2.imwrite(filename,img[y1:y1+y2,x1:x1+x2])
    plt.show()
```
